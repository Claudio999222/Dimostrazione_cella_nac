```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import backend as K
```

#### Creo il mio dataset tridimensionale e lo trasformo in un array 

```python
x_range = np.arange(-100, 100, 0.2)
y_range = np.arange(-100, 100, 0.2)

res = []

for i, x in enumerate(x_range):
    for l, y in enumerate(y_range):
        res.append([x, y, K.sigmoid(x) * K.tanh(y)])

res = np.asarray(res, dtype = 'float32')
```

#### Come possiamo vedere i pesi che usciranno dalla nac saranno valori di -1, 0, 1 se non i rarissimi casi diversi ma sempre nel range

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(res[:, 0], res[:, 1], res[:, 2], alpha=0.1, zdir='z', s=0.1)
plt.show()
```

```python
x_range = np.arange(-2, 2, 0.1)
y_range = np.arange(-2, 2, 0.1)

res = []

for i, x in enumerate(x_range):
    for l, y in enumerate(y_range):
        res.append([x, y, K.tanh(x) * K.sigmoid(y)])

res = np.asarray(res, dtype='float32')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(res[:, 0], res[:, 1], res[:, 2], s=1, alpha=0.25)
```
